```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('titanic-passengers.csv' , sep=";")
df
```

```python
df.isnull()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df
```

```python
ne=len(df['Cabin'])
print("nbr elements:",ne)
print(df["Cabin"].value_counts()) #G6 is 691 times / 890 !
df['Cabin'].fillna('G6', inplace=True)
df
```

```python
g=sns.FacetGrid(df, row="Sex",col='Survived',height=2,aspect=1.5)
g.map(sns.barplot,'Age', alpha=0.5, ci=None)
g.add_legend()
```

```python
from sklearn.preprocessing import LabelEncoder
enc= LabelEncoder()
df["Survived"]=enc.fit_transform(df["Survived"])
g=sns.FacetGrid(df,col='Survived')
g.map(plt.hist,'Pclass', bins=20)
g.add_legend()
```

```python
g=sns.FacetGrid(df,col='Survived')
g.map(plt.hist,'SibSp', bins=20)
g.add_legend()
```

```python
df=pd.read_csv('titanic-passengers.csv',sep=";")
def plot_correlation_map(df):
    corr=df.corr()
    s,ax=plt.subplots(figsize=(13,11))
    cmap=sns.diverging_palette(220,10,as_cmap=True)
    s=sns.heatmap(
        corr,
        cmap=cmap,
        square=True,
        cbar_kws={'shrink': .9},
        ax=ax,
        annot=True,
        annot_kws={'fontsize':12}
    )
plot_correlation_map(df)
```

```python
df=pd.read_csv('titanic-passengers.csv',sep=";")
titled=[name[name.find(',')+2: name.find('.')] for name in df['Name']]
df['Title']=pd.DataFrame(titled)
df['FamilySize']=df['SibSp']+df['Parch']
df.drop(["Name","SibSp","Parch"],axis=1,inplace=True)
df
```

```python
def plot_correlation_map(df):
    corr=df.corr()
    s,ax=plt.subplots(figsize=(13,11))
    cmap=sns.diverging_palette(220,10,as_cmap=True)
    s=sns.heatmap(
        corr,
        cmap=cmap,
        square=True,
        cbar_kws={'shrink': .9},
        ax=ax,
        annot=True,
        annot_kws={'fontsize':12}
    )
plot_correlation_map(df)
```
